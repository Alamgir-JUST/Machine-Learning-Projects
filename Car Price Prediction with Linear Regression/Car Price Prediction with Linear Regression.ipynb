```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing Dataset
df = pd.read_csv('carprices.csv')
```

```python
df
```

```python
#Features: Car Model, Mileage,Age:: Label: Sell Price
df.shape
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

# Dummy data chacking
# Car Model data is string, applying one hot encoding to transfer into numerical data

```python
dummies = pd.get_dummies(df['Car Model'])
```

```python
dummies
```

# Mering the new dataset with the previous dataset

```python
merged = pd.concat([df, dummies], axis = 'columns')
```

```python
merged
```

```python
final_dataset = merged.drop(['Car Model'], axis = 'columns')
```

```python
final_dataset
```

# Dividing Columns into Features and Class

```python
y = df['Sell Price']
```

```python
y
```

```python
x = final_dataset.drop(['Sell Price'], axis = 'columns')
```

```python
x
```

# Model Developing

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = .30, random_state = 1)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
#Making Prediction
lr.predict(xtest)
```

```python
ytest
```

```python
accuracy = lr.score(x, y)
```

```python
print ('Model Accuracy is: ', accuracy * 100, '%')
```
