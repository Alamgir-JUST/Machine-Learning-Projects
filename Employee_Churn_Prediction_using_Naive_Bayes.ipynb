# **Complete Coding with Naive Bayes**

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the training and testing datasets
df = pd.read_csv('/content/drive/MyDrive/Machine Learning Lab/employee_churn_dataset.csv')
```

```python
df.value_counts("Churn")
```

```python
# Importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Data Preprocessing
# Encoding categorical features using LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to each feature
data_encoded = df.copy()
for column in data_encoded.columns:
    data_encoded[column] = label_encoder.fit_transform(data_encoded[column])

# Step 2: Splitting data into features (X) and target (y)
X = data_encoded.drop('Churn', axis=1)
y = data_encoded['Churn']

# Step 3: Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train Naive Bayes model
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)

# Step 5: Make predictions on the test set
y_pred = nb_classifier.predict(X_test)

# Step 6: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Display results
print(f'Accuracy: {accuracy * 100:.2f}%')
print('Classification Report:')
print(report)
```
